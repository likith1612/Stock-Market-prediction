```python
#stock price
import pandas as pd
import numpy as np

df = pd.read_csv("/Stock market prediction/NSE-Tata-Global-Beverages-Limited.csv")
df.head()

x = df[["Open"]]
y = df[["Close"]]

x.shape

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
x_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train = x_scaler.fit_transform(np.array(x_train).reshape(-1,1))
x_test = x_scaler.transform(np.array(x_test).reshape(-1,1))
y_train = y_scaler.fit_transform(np.array(y_train).reshape(-1,1))
y_test = y_scaler.transform(np.array(y_test).reshape(-1,1))

x_train.shape,x_test.shape,y_train.shape,

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Flatten

model = Sequential()
model.add(LSTM(128,return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(64,return_sequences=False))
model.add(Dense(64))
model.add(Dense(32))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mean_squared_error',metrics='mse')
model.fit(x_train,y_train,epochs=20)

model.evaluate(x_test,y_test)

op = model.predict(x_scaler.transform(np.array([208]).reshape(-1,1)))
y_scaler.inverse_transform(op)
```
